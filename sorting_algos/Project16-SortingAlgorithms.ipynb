```
# Importing numpy so we can use numpy arrays and generate random data
import numpy as np
from time import time
import matplotlib.pyplot as plt


# Making a convenient function for producing scrambled lists of given length
def make_scrambled_data(length):
    """Returns an array of size 'length' of random numbers between 0 and 100
    """
    return np.random.randint(0, 100, length)


def check_sorting_algo(sorting_func, verbose=True):
    """Tests the given sorting function on a 1000 length array and returns True
    if the array is sorted or False if not. Prints the array before and after
    unless verbose=False
    """
    scrambled = make_scrambled_data(10)
    if verbose:
        print(f'Original: {scrambled}')
    test_sorted = np.array(sorting_func(scrambled))
    is_sorted = (test_sorted == np.sort(test_sorted)).all()
    if verbose:
        print(f'Sorted:   {test_sorted}')
    return is_sorted


def time_sorting_algo(sorting_func, array_length, n_trials=3):
    """Tests the given sorting function on a 1000 length array 'n_trials' times
    and returns the mean time it took to sort the array.
    """
    durations = []
    for i in range(n_trials):
        scrambled = make_scrambled_data(array_length)
        start = time()
        test_sorted = sorting_func(scrambled)
        if not test_sorted:
          print("The sorting algo didn't work properly.")
        duration = time() - start
        durations.append(duration)
    return np.mean(durations)
```

### Modeling Time Complexity with Mathematical Functions
The natural way to describe time complexity is to model algorithm run time ($O$) as a function of array length ($n$). Using [`curve_fit`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html) from scipy, try to fit each of the sorting algorithms with an appropriate function from the following options (in order of fastest to slowest): 
1. $O(1)$ $\rightarrow$ (*fastest category*: the algorithm takes roughly the same time regardless of the array size) 
1. $O($log $n)$
1. $O(n)$ $\rightarrow$ (We can think of this as a single for loop)
1. $O(n\cdot$log $n)$
1. $O(n^b)$
1. $O(b^n)$ $\rightarrow$ (*slowest category*: the algorithm's run time scales exponentially with the array size)

This notation is called **big-O** and it acts as a rating system which we use to categorize algorithms by the general form of their time complexity. $O(f(n))$ is shorthand for any function $a\cdot f(n)$, where $a$ is a constant (which are we'll be fitting with `curve_fit`) We may need to also add an additional constant parameter, $b$, (for 4. and 5.) but that is shown in the list above.

For example, if the runtimes for `bubbleSort` fit the relationship $4n^2$ or $0.324n^2$ or $9999n^2$, we would simply say that `bubbleSort` scales as $O(n^2)$ because the $a$ isn't that important compared to the $n^2$ since we really just want to know what the general shape of the relationship is.

### Watch a Video Tutorial
If you want a more in-depth description of **Big O Notation** check out this [awesome video](https://www.youtube.com/watch?v=D6xkbGLQesk) from CSDojo (it's a bit long but it's fairly understandable and still thorough).

```
from scipy.optimize import curve_fit

model_lengths = np.arange(lengths[0], lengths[-1], 10)

### Example of testing bubbleSort, timSort (Python built-in algorithm) with 
### some of the functions

def O_1(x, a):
  return [a]*len(x)

def O_logn(x, a):
  return a*np.log(x)

def O_bn(x, a, b):
  return a*b**x

algo_list = ['bubbleSort']
times_list = [bubble_times]

func_list = [O_1, O_logn, O_bn]

for times, algo in zip(times_list, algo_list):
  plt.plot(lengths, times, 'ko-', label='data')
  for func in func_list:
    popt, pcov = curve_fit(func, lengths, times)
    model_times = func(model_lengths, *popt)
    plt.plot(model_lengths, model_times, label=f'fit: {func.__name__}')
    print(f'{func.__name__} fitted parameters = {popt}') # looks like [a] or [a,b]
    print(f'{func.__name__} variance = {np.diag(pcov)}') # looks like [a] or [a,b]
  plt.title(algo)
  plt.legend()
  plt.show()
```
